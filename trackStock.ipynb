我們會生成一個物件，用來協助我們追蹤在追蹤清單上的股票，這樣有幾項好處：
1. 不用一直重複抓取歷史資料
2. 可依據個別股票作個性化追蹤


```python
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
import math
import plotly.graph_objects as go
```

```python
class trackStock:
    '''
    Create a class to track historical data the stock 
    '''
    def __init__(self, symbol):
        self.symbol = symbol
        self.start_time = math.floor((datetime.datetime.now() - relativedelta(years=1)).timestamp())
        self.end_time = math.floor(datetime.datetime.now().timestamp())
        self.history = None
    
    def fetch(self):
        self.start_time = math.floor((datetime.datetime.now() - relativedelta(years=1)).timestamp())
        self.end_time = math.floor(datetime.datetime.now().timestamp())
        self.history = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/{}?period1={}&period2={}&interval=1d&events=history'\
                                   .format(self.symbol, self.start_time, self.end_time))
```
